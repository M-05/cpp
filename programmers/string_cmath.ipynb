# [문자열의 앞의 n글자](https://school.programmers.co.kr/learn/courses/30/lessons/181907)

```python
#include <string>
#include <vector>
#include <iostream>
using namespace std;

string solution(string my_string, int n) {
     string answer;
     for (int i=0; i < n; i++)
     {
         answer += my_string[i];
     }
     return answer;
}
```

```python
#include <string>
#include <vector>
#include <iostream>
using namespace std;

string solution(string my_string, int n) {
    return my_string.substr(0, n);
}
```

```python
#include <string>
#include <vector>

using namespace std;

string solution(string my_string, int n) {
    string answer = "";
    answer.assign(my_string.begin(), my_string.begin()+n);
    return answer;
}
```
